```python
from pathlib import Path
import pandas as pd
```

```python
def join_city_demand_weather(demand_file: str, weather_file: str) -> pd.DataFrame:
    demand = pd.read_csv("D:/Projects/chicago-bike-demand-intelligence/02_data_processed/city_hourly_demand.csv")
    weather = pd.read_csv("D:/Projects/chicago-bike-demand-intelligence/02_data_processed/weather_hourly.csv")

    # Make sure join keys match
    demand["date"] = pd.to_datetime(demand["date"]).dt.date
    weather["date"] = pd.to_datetime(weather["date"]).dt.date

    demand["hour"] = demand["hour"].astype(int)
    weather["hour"] = weather["hour"].astype(int)

    # Left join so we keep all demand rows
    merged = demand.merge(
        weather[["date", "hour", "temperature", "precipitation_mm", "wind_speed"]],
        on=["date", "hour"],
        how="left"
    )

    return merged


if __name__ == "__main__":
    demand_file = "D:/Projects/chicago-bike-demand-intelligence/02_data_processed/city_hourly_demand.csv"
    weather_file = "D:/Projects/chicago-bike-demand-intelligence/02_data_processed/weather_hourly.csv"

    output_path = Path("D:/Projects/chicago-bike-demand-intelligence/02_data_processed/city_hourly_demand_weather.csv")
    output_path.parent.mkdir(parents=True, exist_ok=True)

    merged_df = join_city_demand_weather(demand_file, weather_file)
    merged_df.to_csv(output_path, index=False)

    print("City demand + weather file created")
    print(merged_df.head())
    print(merged_df.shape)
    print(merged_df.isna().sum())
```
