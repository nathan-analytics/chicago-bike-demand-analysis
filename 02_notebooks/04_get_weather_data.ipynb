```python
import requests
import pandas as pd
from pathlib import Path
```

```python
def get_weather_data(start_date: str, end_date: str):
    url = "https://archive-api.open-meteo.com/v1/archive"

    params = {
        "latitude": 41.8781,   # Chicago
        "longitude": -87.6298,
        "start_date": start_date,
        "end_date": end_date,
        "hourly": [
            "temperature_2m",
            "precipitation",
            "windspeed_10m"
        ],
        "timezone": "America/Chicago"
    }

    response = requests.get(url, params=params)
    data = response.json()

    df = pd.DataFrame(data["hourly"])

    return df


if __name__ == "__main__":
    # Match your dataset range (adjust if needed)
    start_date = "2025-03-01"
    end_date = "2025-05-31"

    df_weather = get_weather_data(start_date, end_date)

    # Convert time column
    df_weather["time"] = pd.to_datetime(df_weather["time"])

    # Create join keys
    df_weather["date"] = df_weather["time"].dt.date
    df_weather["hour"] = df_weather["time"].dt.hour

    # Rename columns (clean)
    df_weather = df_weather.rename(columns={
        "temperature_2m": "temperature",
        "precipitation": "precipitation_mm",
        "windspeed_10m": "wind_speed"
    })

    # Save
    output_path = Path("D:/Projects/chicago-bike-demand-intelligence/02_data_processed/weather_hourly.csv")
    output_path.parent.mkdir(parents=True, exist_ok=True)

    df_weather.to_csv(output_path, index=False)

    print("Weather data saved")
    print(df_weather.head())
    print(df_weather.shape)
```
